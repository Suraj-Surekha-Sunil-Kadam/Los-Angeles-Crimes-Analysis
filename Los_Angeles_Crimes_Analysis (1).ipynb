```python
# Demographic Influence Hypothesis

import scipy.stats as stats

# creating the contingency table for Victim Descent vs. Crime Type
contingency_table = pd.crosstab(data['Vict Descent'], data['Crm Cd Desc'])

# performing the chi-square test
chi2_stat, p, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p}")
# print(f"Degrees of Freedom: {dof}")
# print(f"Expected Frequencies: {expected}")
```

```python
# Temporal Crime Occurrence Hypothesis

# Ensure TIME OCC is in proper datetime format
data['TIME OCC'] = pd.to_datetime(data['TIME OCC'], format='%H:%M:%S')

# Extract hour from TIME OCC
data['Hour OCC'] = data['TIME OCC'].dt.hour  # ✅ Convert datetime to numerical hour

# Define time blocks
bins = [0, 6, 12, 18, 24]
labels = ['Midnight - Morning', 'Morning - Noon', 'Afternoon - Evening', 'Evening - Night']
data['Time Block'] = pd.cut(data['Hour OCC'], bins=bins, labels=labels, right=False)  # ✅ Now it works!

# Group data by time block
crime_time_counts = data.groupby('Time Block')['DR_NO'].count()

# Perform One-Way ANOVA
anova_result = stats.f_oneway(
    data[data['Time Block'] == 'Midnight - Morning']['Hour OCC'],
    data[data['Time Block'] == 'Morning - Noon']['Hour OCC'],
    data[data['Time Block'] == 'Afternoon - Evening']['Hour OCC'],
    data[data['Time Block'] == 'Evening - Night']['Hour OCC']
)

# Print ANOVA results
print(f"ANOVA Statistic: {anova_result.statistic}")
print(f"P-value: {anova_result.pvalue}")

# Interpretation
if anova_result.pvalue < 0.05:
    print("Accept H1: Specific time blocks (hours of the day) have statistically significant higher crime rates")
else:
    print("Reject H₀: Crime occurrence is random across time blocks.")
```

```python

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=crime_time_counts.index, y=crime_time_counts.values, palette="coolwarm")
plt.title('Crime Occurrences by Time Block')
plt.xlabel('Time Block')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
# @title
# Spatial Distribution Hypothesis

# Count occurrences of crime in each area
crime_area_counts = data['AREA NAME'].value_counts()

# Expected distribution assuming uniformity
expected_counts = [crime_area_counts.mean()] * len(crime_area_counts)

# Perform Chi-Square Goodness of Fit test
chi2_stat, p = stats.chisquare(crime_area_counts, expected_counts)

print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p}")

# Interpretation
if p < 0.05:
    print("Accept H1: Certain areas have significantly higher crime rates or specific crime type concentrations.")
else:
    print("Reject H0: Crime is uniformly distributed across different areas.")
```
